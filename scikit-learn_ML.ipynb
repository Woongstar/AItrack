```
# 라이브러리 호출

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```
# Dataset 호출
import pandas as pd

iris = load_iris()
```

```
iris.keys()
```

```
iris['filename']
```

```
iris_data = iris.data
print(type(iris_data))
iris_label = iris.target

print('iris target:', iris_label)
print('iris target names:{}'. format(iris.target_names))

iris_df = pd.DataFrame(data= iris_data, columns = iris.feature_names)
iris_df['label'] = iris.target
iris_df.head(3)
```

```
# DataFrame 구성
```

```
#Data split
X_train,X_test,y_train,y_test = train_test_split(iris_data,iris_label, test_size=0.2, random_state=123)
```

```
# Data shape  확인
print(X_train.shape)
print(X_test.shape)
```

```
#%whos 로 생성된 변수 알아보기
%whos
```

```
#DataFramework info()
iris_df.info()
```

```
#DataFramework describe()
iris_df.describe()
```

```
#알고리즘 생성 및 학습
#DecisionTreeClassifier 생성 및 학습

dt_clf = DecisionTreeClassifier(random_state=11)

dt_clf.fit(X_train,y_train)
```

```
#Save Model
#변수 저장
import pickle
#파일 저장
from sklearn.externals import joblib 

saved_model = pickle.dumps(dt_clf)

pickle_pred = pickle.loads(saved_model)

pred = pickle_pred.predict(X_test)
```

```
#Load Model
joblib.dump(dt_clf,'iris_dump.pkl')
pred1 = joblib.load('iris_dump.pkl')

pred1 = pred1.predict(X_test)
```

```
#학습이 완료된 모델에 테스트 테이터로 예측 수행
pred2 = dt_clf.predict(X_test)

from sklearn.metrics import accuracy_score

result = accuracy_score(y_test, pred)
print ('accuracy_score:{}', format(result))

result = accuracy_score(y_test, pred1)
print ('accuracy_score2:{}', format(result))

result = accuracy_score(y_test, pred2)
print ('accuracy_score3:{}', format(result))
```
